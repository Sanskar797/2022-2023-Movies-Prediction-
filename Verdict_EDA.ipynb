#### A function to explore some columns

```python
def rating(col,n):
    if n==min:
        s=movies[col].min()
        t=movies.loc[movies[col] == s]
        return s,t
    elif n==max:
        p=movies[col].max()
        t=movies.loc[movies[col] == p]
        return p,t
```

### EXPLORING THE RELATIONSHIP BETWEEN VERDICT AND INDEPENDENT VARIABLES

```python
## Functions to Explore the relationship between Verdict and categorical independent variables

def cross_tabs(col_name):
    a = pd.crosstab(movies['Verdict'], movies[col_name])
    return a

def cat_plots(col_name):
    plt.figure(figsize=(10,3))
    sns.countplot(x="Verdict", hue=col_name, data=movies)
    plt.title(f"Distribution of Verdict by {col_name}")
    plt.xlabel('Verdict')
    plt.ylabel('Count')
    plt.legend(loc='upper right')
    plt.show()
```
